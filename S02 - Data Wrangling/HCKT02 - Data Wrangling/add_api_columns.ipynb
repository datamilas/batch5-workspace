```python
def extractAPIdata(id_list, features_list=None):
    base_url = 'https://xauengvo49.execute-api.eu-west-1.amazonaws.com/dev/missingdata/'
    
    dict_list = []
    
    for i in range(len(id_list)):
        response = requests.get(base_url+str(id_list[i]))
        feature_values = response.json()
        feature_values['id'] = id_list[i]
        dict_list.append(feature_values)
        
    df = pd.DataFrame.from_records(dict_list)
    df = df.set_index('id')
    
    
    return df
    
```
